```python
# Pandas for data manipulation
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

# Read csv file and define dataframe object (df for dataframe)
df = pd.read_csv('./data/gapminder_world_development.txt', sep='\t')

df.head()  # show dataframe header to stdout
```

```python
print(df.dtypes)
print(df.describe())
```

```python
df["lifeExp"].mean()
```

```python
plt.hist(df["lifeExp"])
plt.show()
```

```python
plt.hist(df["lifeExp"],bins=6)
plt.show()
```

```python
gm1900 = df[df.year == 1900]
gm2015 = df[df.year == 2015]
```

```python
plt.hist(gm2015["lifeExp"],bins=12)
plt.show()
```

```python
plt.hist(gm1900["lifeExp"],bins=12)
plt.show()
```

```python
plt.scatter(df.year,df.lifeExp)
plt.show()
```

```python
lifeExpTitle = "Life Expectancy [in years]"
gdpTitle = "GDP per Capita [in USD]"


plt.scatter(gm2015.gdp,gm2015.lifeExp,)
plt.ylabel(lifeExpTitle)
plt.xlabel(gdpTitle)
plt.title("World Development in 2015")
plt.yticks([0,30,60,90])
plt.show()
```

```python
plt.scatter(gm1900.gdp,gm1900.lifeExp)
plt.ylabel(lifeExpTitle)
plt.xlabel(gdpTitle)
plt.title("World Development in 1900")
plt.yticks([0,30,60,90])
plt.show()
```

```python
pop2015 = np.array(gm2015["pop"])
pop1900 = np.array(gm1900["pop"])
```

```python
pop2015 = pop2015 / 1000000
pop1900 = pop1900 / 1000000
```

```python
plt.scatter(gm2015.gdp,gm2015.lifeExp,s=pop2015,alpha=0.6)
plt.ylabel(lifeExpTitle)
plt.xlabel(gdpTitle)
plt.title("World Development in 2015")
plt.yticks([0,30,60,90])
plt.xscale('log')
#plt.grid(True)
plt.show()
```

```python
plt.scatter(gm1900.gdp,gm1900.lifeExp,s=pop1900)
plt.xlabel(lifeExpTitle)
plt.ylabel(gdpTitle)
plt.title("World Development in 1900")
plt.yticks([0,30,60,90])
plt.xscale('log')
plt.show()
```
