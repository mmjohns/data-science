```python
# Built-in python libraries
import sys
import os

# 3rd-party libraries
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
degree_data = pd.read_csv("./data/degrees_gender_long.csv")
degree_data.head()
```

```python
degrees_total = degree_data.groupby(['Gender'])['Num'].sum().reset_index()
degrees_total
```

```python
# Add Columns
degrees_total['DegreeType'] = 'Total'

# Get percentages
degrees_total['Num'] = degrees_total['Num'] / degrees_total['Num'].sum()
degrees_total
```

```python
degrees_types = degree_data.groupby(['DegreeType', 'Gender'])['Num'].sum()
degrees_types
```

```python
# Get percentages
degrees_types = degrees_types.groupby(level=0).apply(lambda x: 100 * x / float(x.sum())).reset_index()
degrees_types
```

Seaborn

```python
import seaborn as sns
```

```python
def plot_v1(data):
    # Create the bar plot
    ax = sns.barplot(
        x="DegreeType", y="Num", hue="Gender",
        order=["Certificate", "Master", "Doctoral"],
        hue_order=["Male", "Female"],
        data=data)

    # Plot the 50% line
    ax.hlines(50, -0.5, 4.5, linestyle='--', linewidth=1)
    ax.set_ylim(0, 100)
    
    # Return the figure object and axis
    return plt.gcf(), ax


plot_v1(degrees_types);
```

Add/Change Labeling

```python
def plot_v2(data):
    # Specify that I want each subplot to correspond to
    # a different robot type
    g = sns.FacetGrid(
        data,
        col="DegreeType",
        col_order=["Certificate", "Master", "Doctoral"],
        sharex=False)

    # Create the bar plot on each subplot
    g.map(
        sns.barplot,
        "DegreeType", "Num", "Gender",
        hue_order=["Male", "Female"])

    # Now I need to draw the 50% lines on each subplot
    # separately
    axes = np.array(g.axes.flat)
    for ax in axes:
        ax.hlines(50, -0.5, 0.5, linestyle='--', linewidth=1)
        ax.set_ylim(0, 100)

    # Return the figure and axes objects
    return plt.gcf(), axes


plot_v2(degrees_types);
```

```python
def set_labels(fig, axes):
    # These are the labels of each subplot
    labels = ["Certificate", "Master", "Doctoral"]
    
    # Iterate over each subplot and set the labels
    for i, ax in enumerate(axes):

        # Set the x-axis ticklabels
        ax.set_xticks([-.2, .2])
        ax.set_xticklabels(["Male", "Female"])

        # Set the label for each subplot
        ax.set_xlabel(labels[i])
        
        # Remove the y-axis label and title
        ax.set_ylabel("")
        ax.set_title("")
    
    # Set the y-axis label only for the left subplot
    axes.flat[0].set_ylabel("% of degrees awarded")
    
    # Remove the "spines" (the lines surrounding the subplot)
    # including the left spine for the 2nd and 3rd subplots
    sns.despine(ax=axes[1], left=True)
    sns.despine(ax=axes[2], left=True)

    # Set the overall title for the plot
    fig.suptitle("Degrees Awarded by Gender", fontsize=12, x=0.55)


fig, axes = plot_v2(degrees_types)
set_labels(fig, axes)
```

Set styles

```python
def set_style():
    # This sets reasonable defaults for font size for
    # a figure that will go in a paper
    sns.set_context("paper")
    
    # Set the font to be serif, rather than sans
    sns.set(font='sans')
    
    # Make the background white, and specify the
    # specific font family
    sns.set_style("white", {
        "font.family": "sans",
        "font.sans": ["Roboto", "Arial", "sans"]
    })


set_style()
fig, axes = plot_v2(degrees_types)
set_labels(fig, axes)
```

Colors

```python
def get_colors():
    return sns.color_palette("Paired")
```

```python
sns.palplot(get_colors())
```

```python
def color_bars(axes, colors):
    # Iterate over each subplot
    for i in range(3):

        # Pull out the dark and light colors for
        # the current subplot
        dark_color = colors[i*2 + 1]
        light_color = colors[i*2]

        # These are the patches (matplotlib's terminology
        # for the rectangles corresponding to the bars)
        p1, p2 = axes[i].patches

        # The first bar gets the dark color
        p2.set_color(dark_color)
        
        # The second bar gets the light color, plus
        # hatch marks int he dark color
        p1.set_color(light_color)
        p1.set_edgecolor(dark_color)
        p1.set_hatch('////')


set_style()
fig, axes = plot_v2(degrees_types)
set_labels(fig, axes)
color_bars(axes, get_colors())
```

Plot Sizing

```python
def set_size(fig):
    fig.set_size_inches(8, 4)
    plt.tight_layout()
    
set_style()
fig, axes = plot_v2(degrees_types)
set_labels(fig, axes)
color_bars(axes, get_colors())
set_size(fig)
```
